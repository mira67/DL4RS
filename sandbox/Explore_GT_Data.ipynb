#Q1: How the ice fraction is related to melt pond fraction?
1. If there are both open water and melt pond, ice fraction is not linear to melt pond, linear to openwater. 
2. If there is only melt pond and ice, then ice fraction is sure to be linear with melt pond fraction.

```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.rcParams['figure.figsize'] = (10,10) # Make the figures a bit bigger

tf = "/home/mirabot/googledrive/deeplearning/seaiceprj/G02159_Pond_statistics_spreadsheets/cacana_2000_irt_stats.xls"
table = pd.read_excel(tf,
                      sheetname = 'cacana_2000aug15a_3c_stats_v1.0',
                      header = 0,
                      parse_cols = "B,C,D,E",
                      convert_float = False)
#print(table)
table = table.replace(9.99, np.nan)
row_index = table.set_index(np.linspace(0, 400, num=400))
plt.figure(); table.plot(subplots=True, figsize=(12, 12)); plt.legend(loc='best')
scatter_matrix(table, alpha=0.2, figsize=(12, 12), diagonal='kde')
```
